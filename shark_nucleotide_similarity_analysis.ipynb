```python
# Gerekli Kütüphanelerin İçe Aktarılması :
import pandas as pd
import numpy as np
from Bio import SeqIO
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
from scipy.cluster.hierarchy import linkage, fcluster, dendrogram
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score



# Grafiklerin daha güzel görünmesi için seaborn stilini ayarlama :
sns.set(style="whitegrid")
```

```python
# Canlıların FASTA Dosyalarının dosyalarının dictionary yapısında tutulması :
fasta_files_dict = {
    'Shark': [
        "data/shark/shark_sequence1.fasta", 
        "data/shark/shark_sequence2.fasta"
    ],
    'Turtle': [
        "data/turtle/turtle_sequence1.fasta", 
        "data/turtle/turtle_sequence2.fasta"
    ],
    'Mouse': [
        "data/mouse/mouse_sequence1.fasta", 
        "data/mouse/mouse_sequence2.fasta"
    ],
    'Dolphin': [
        "data/dolphin/dolphin_sequence1.fasta", 
        "data/dolphin/dolphin_sequence2.fasta"
    ],
    'Blue Whale': [
        "data/blue_whale/blue_whale_sequence1.fasta", 
        "data/blue_whale/blue_whale_sequence2.fasta"
    ]
}
```

```python
# Veri Okuma Fonksiyonunun Tanımlanması :
# Belirtilen FASTA dosyalarını okuyup bir DataFrame'e dönüştürür.
def read_fasta_files(fasta_files, species_name):
    sequences = []
    labels = []
    for file in fasta_files:
        try:
            # Her bir FASTA dosyası okunup sekansı ve etiketi listelere eklenir
            record = SeqIO.read(file, "fasta")
            sequences.append(str(record.seq))
            labels.append(record.id)
        except Exception as e:
            print(f"Error reading {file}: {e}")
    # Verileri bir DataFrame'e dönüştürelim
    df = pd.DataFrame({
        'Sequence': sequences,
        'Species': [species_name] * len(sequences) 
    })
    return df
```

```python
# Veri Okuma ve tüm canlıların verilerini dataframe'e dönüştürme :
all_dfs = []

for species, files in fasta_files_dict.items():
    df = read_fasta_files(files, species)
    all_dfs.append(df)

combined_df = pd.concat(all_dfs, ignore_index=True)

print("Combined DataFrame:")
print(combined_df)
```

```python
# Veri Temizleme Fonksiyonunun Tanımlanması : (Veri ön işleme adımı)
# DNA sekanslarında boşluk, geçersiz karakterler gibi hatalı verileri olabilir.Bu fonksiyon bu hatalı verileri temizler.
# Ayrıca nükleotit dizisinde "N" herhangi bir nükleotidi temsil eder ve eğer varsa bu karakterleri de temizler.
def clean_sequence(seq):
    valid_nucleotides = set('ATGC')  # Geçerli nükleotitler
    # Sekanstaki her bir karakteri kontrol edip geçerli olanları birleştiriyoruz
    cleaned_seq = ''.join([nuc for nuc in seq.upper() if nuc in valid_nucleotides])
    return cleaned_seq
```

```python
# Veri Temizleme ve Temizlenmiş Sekansların Görüntülenmesi :
combined_df['Cleaned_Sequence'] = combined_df['Sequence'].apply(clean_sequence)

print("\nTemizlenmiş Sekansların İlk 5 Satırı:")
print(combined_df[['Sequence', 'Cleaned_Sequence', 'Species']].head())
```

```python
# Sekans Uzunluklarının Hesaplanması ve İstatistiklerin Görüntülenmesi :
combined_df['Sequence_Length'] = combined_df['Cleaned_Sequence'].apply(len)

print("\nSekans Uzunluklarının İstatistikleri:")
print(combined_df['Sequence_Length'].describe())
```

```python
# Nükleotitlerin Sayısallaştırılması İşlemi için One-Hot Encoding Fonksiyonunun Tanımlanması :
# DNA sekansını sayısal değerlere dönüştürür ve daha kolay işlem yapılmasını sağlar.
def one_hot_encode(sequence):
    mapping = {'A':0, 'T':1, 'G':2, 'C':3}  # Nükleotit haritası

    # Her nükleotiti sayısal değere dönüştür, geçersiz karakterler için -1 ataması yapar.
    encoded = [mapping.get(nuc, -1) for nuc in sequence]
    return encoded
```

```python
# Nükleotitlerin Sayısallaştırılması ve Maksimum Sekans Uzunluğunun Belirlenmesi :
combined_df['Encoded_Sequence'] = combined_df['Cleaned_Sequence'].apply(one_hot_encode)

# Maksimum sekans uzunluğunu bulma :
max_length = combined_df['Encoded_Sequence'].apply(len).max()
print(f"\nMaksimum Sekans Uzunluğu: {max_length}")
```

```python
# Sekansların Aynı Uzunlukta Pad Edilmesi :
# Bu işlem, makine öğrenimi algoritmaları için giriş verilerinin aynı uzunlukta olmasını sağlar.
def pad_sequence(seq, max_len):
    return seq + [-1]*(max_len - len(seq))  # Eksik kısımları -1 ile doldur

# Pad edilmiş sekansları ekleme :
combined_df['Padded_Sequence'] = combined_df['Encoded_Sequence'].apply(lambda x: pad_sequence(x, max_length))

# İlk birkaç pad edilmiş sekansı kontrol edelim
print("\nPad edilmiş sekansların ilk 5 satırı:")
print(combined_df[['Encoded_Sequence', 'Padded_Sequence']].head())
```

```python
# Veri Matrisinin Oluşturulması ve PCA ile Boyut İndirgeme :
# Veri matrisi oluşturma :
X = np.array(combined_df['Padded_Sequence'].tolist())  # Pad edilmiş sekanslar
y = combined_df['Species']  # Tür etiketleri

# PCA ile boyut indirgeme :
pca = PCA(n_components=2, random_state=42)  # 2 bileşen seçiyoruz
X_pca = pca.fit_transform(X)

# PCA sonuçlarını DataFrame'e ekleme :
combined_df['PCA1'] = X_pca[:,0]
combined_df['PCA2'] = X_pca[:,1]

print("\nPCA DataFrame:")
print(combined_df[['PCA1', 'PCA2', 'Species']].head())
```

```python
# Elbow Yöntemi ile Optimal Küme Sayısını Belirleme :
# Burada bir dirsek grafiği çizilir ve dirseğin olduğu nokta, optimal küme sayısını belirler.
# Bu noktadan sonra daha fazla küme eklemek, modelin performansını artırmaz.
# Bizim grafiğimize göre küme sayısını "4" olarak belirleyebiliriz.


# Elbow Yöntemi ile optimal küme sayısını belirleme
sse = {}
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_pca)
    sse[k] = kmeans.inertia_

# Elbow grafiğini çizelim
plt.figure(figsize=(10,6))
plt.plot(list(sse.keys()), list(sse.values()), marker='o')
plt.title('Elbow Yöntemi ile Optimal Küme Sayısının Belirlenmesi')
plt.xlabel('Küme Sayısı (K)')
plt.ylabel('SSE (Sum of Squared Errors)')
plt.xticks(range(2, 10))
plt.grid(True)
plt.show()
```

```python
k_optimal = 3  # Elbow yöntemi ile belirlenen optimal küme sayısı

kmeans = KMeans(n_clusters=k_optimal, random_state=42).fit(X_pca)
cluster_labels = kmeans.labels_

# Metriklerin Hesaplanması
sil_score = silhouette_score(X_pca, cluster_labels)
ch_score = calinski_harabasz_score(X_pca, cluster_labels)
db_score = davies_bouldin_score(X_pca, cluster_labels)

print(f"\nK={k_optimal} için K-means Kümeleme Sonuçları:")
print(f"Silhouette Skoru: {sil_score:.2f}")
print(f"Calinski-Harabasz Skoru: {ch_score:.2f}")
print(f"Davies-Bouldin Skoru: {db_score:.2f}")


combined_df['Cluster'] = cluster_labels

# Kümeleme Sonuçlarının Görselleştirilmesi
plt.figure(figsize=(10, 8))
sns.scatterplot(
    x='PCA1',
    y='PCA2',
    hue='Cluster',
    data=combined_df,
    palette='Set2',
    s=100
)
plt.title(f'K-Means Kümeleme Sonuçları (K=4, PCA ile Boyut İndirgeme)')
plt.xlabel('PCA Bileşen 1')
plt.ylabel('PCA Bileşen 2')
plt.legend(title='Kümeler', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
k = 4  # Küme sayısı

linkage_matrix = linkage(X_pca, method='ward')  # 'ward' bağlantı yöntemi kullanıldı

scores_hierarchical = []


# Hiyerarşik Kümeleme işlemi
hierarchical_labels = fcluster(linkage_matrix, t=k, criterion='maxclust')

# Skor hesaplama
sil_score = silhouette_score(X_pca, hierarchical_labels)
ch_score = calinski_harabasz_score(X_pca, hierarchical_labels)
db_score = davies_bouldin_score(X_pca, hierarchical_labels)

# Skorları ekleme
scores_hierarchical.append({
    'K': k,
    'Silhouette Skoru': sil_score,
    'Calinski-Harabasz Skoru': ch_score,
    'Davies-Bouldin Skoru': db_score
})

print(f"\nK={k} için Hiyerarşik Kümeleme Sonuçları:")
print(f"Silhouette Skoru: {sil_score:.2f}")
print(f"Calinski-Harabasz Skoru: {ch_score:.2f}")
print(f"Davies-Bouldin Skoru: {db_score:.2f}")

# Dendrogram görselleştirme
plt.figure(figsize=(10, 8))
dendrogram(linkage_matrix, truncate_mode='lastp', p=10, leaf_rotation=45, leaf_font_size=10, show_contracted=True)
plt.title('Hiyerarşik Kümeleme - Dendrogram')
plt.xlabel('Veri Noktaları veya Kümeler')
plt.ylabel('Mesafe')
plt.show()
```

```python
# Benzerlik Fonksiyonunun Tanımlanması :
# İki DNA sekansı arasındaki benzerlik oranını hesaplar.
def similarity(seq1, seq2):
    matches = sum([1 for a, b in zip(seq1, seq2) if a == b])  # Eşleşen nükleotit sayısı
    return matches / len(seq1) if len(seq1) > 0 else 0  # Oranı hesapla
```

```python
# Benzerlik Oranlarının Hesaplanması (Shark'a Göre) :

# Hedef türleri belirleyelim (Shark dahil)
species_list = combined_df['Species'].unique().tolist()

# Sharka ait tüm sekansları seçelim
shark_sequences = combined_df[combined_df['Species'] == 'Shark']['Cleaned_Sequence'].tolist()

all_similarity_scores = []

# Her bir species için benzerlik skorlarını hesaplayalım
for species in species_list:
    # Hedef türden tüm sekansları seçelim
    target_sequences = combined_df[combined_df['Species'] == species]['Cleaned_Sequence'].tolist()
    
    # Her bir target sekansı için tüm Shark sekansları ile benzerlik hesapla
    similarity_scores = []
    for target_seq in target_sequences:
        # Tüm Shark sekansları ile benzerlik hesapla
        seq_similarities = [similarity(shark_seq, target_seq) for shark_seq in shark_sequences]
        # Ortalama benzerlik skorunu al
        avg_similarity = np.mean(seq_similarities) * 100  # Yüzdeye çevirme
        similarity_scores.append(avg_similarity)
    
    # Türün ortalama benzerlik skorunu hesapla (tüm sekanslar için ortalama)
    species_avg_similarity = np.mean(similarity_scores)
    
    # Sonuçları DataFrame'e dönüştür
    similarity_scores_df = pd.DataFrame({
        'Species': [species],
        'Similarity (%)': [species_avg_similarity],
        'Reference_Species': ['Shark']
    })
    
    all_similarity_scores.append(similarity_scores_df)
    
    print(f"\n{species} ile Shark sekansları arasındaki ortalama benzerlik oranı:")
    print(similarity_scores_df)
    
# Tüm benzerlik skorlarını birleştirme
combined_similarity_df = pd.concat(all_similarity_scores, ignore_index=True)

print("\nBenzerlik Oranları:")
print(combined_similarity_df)
```
